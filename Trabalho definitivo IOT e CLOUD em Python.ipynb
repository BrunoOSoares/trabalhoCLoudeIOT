# Índice de Desenvolvimento da Educação do Estado de São Paulo(IDESP)

#### O IDESP é um indicador que avalia a qualidade da escola. Nesta avaliação, considera–se que uma boa escola é aquela em que a maior parte dos alunos aprende as competências e habilidades requeridas para a sua série/ano, num período de tempo ideal – o ano letivo. O IDESP é composto por dois critérios: o desempenho dos alunos nos exames de proficiência do SARESP (o quanto aprenderam) e o fluxo escolar (em quanto tempo aprenderam).

```python
#listas com o tipo de variavel conforme os dados das colunas.

coluna_inteiro = ['ano']
coluna_string = ['diretoria','id_diretoria']
coluna_float = ['nota_idesp_ef_iniciais','nota_idesp_ef_finais','nota_idesp_em']

#Coluna e dados que o usuario deve informar, para que seja feito a consulta no DataFrame.
coluna = input("Digite a coluna: ")
dado = input("Digite o dado: ")

coluna = coluna.lower()
dado = dado.upper()

#condicional para prosseguir na consulta e trazer o dado sem quebra do codigo
if coluna in coluna_inteiro:
    df_definitivo = df.loc[df[coluna] == int(dado)]
    
elif coluna in coluna_string:
    df_definitivo = df.loc[df[coluna] == dado]
    
else:
    df_definitivo = df.loc[df[coluna] == float(dado)]

display(df_definitivo)
```

```python
#Exportação da tabela
arquivo = input("Insira um nome para o arquivo: ")
df_definitivo.to_csv(f"C:/Users/bruno/Downloads/{arquivo}.csv", sep=";" , index = False)
```
